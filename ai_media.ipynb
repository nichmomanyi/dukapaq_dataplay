```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
```

```python
# Set the host IP of the chrome driver
host = "127.0.0.1:8088"



# Set the path of the ChromeDriver executable
s=Service(ChromeDriverManager().install())
# Initialize the chrome driver
options = webdriver.ChromeOptions()
options.add_argument(f"--remote-debugging-address={host}")
options.add_argument("--disable-extensions")
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("--headless")
driver = webdriver.Chrome(service=s)

# Navigate to the login page
driver.get("https://au.agent.ai-media.community/")

# Enter the username and password
username = driver.find_element(By.ID, "username")
password = driver.find_element(By.ID, "password")
username.send_keys("harry.thompson")
password.send_keys("@Mentor254")

# Submit the form
driver.find_element(By.ID, 'submit').click()

# Wait for the page to load
time.sleep(5)

# Navigate to the agent dashboard page
driver.get("https://au.agent.ai-media.community/agent-dashboard")

# Refresh the page every 5 seconds
while True:
    time.sleep(5)
    driver.refresh()
    
    # Check if there's a task available
    try:
        task = driver.find_element_by_xpath('//button[contains(@class, "btn-task")]')
        money = task.find_element_by_xpath('.//span[contains(@class, "money")]')
        print(f"Task available: {money.text}")
        
        # Click the task with the highest amount of money
        task.click()
        break
    except:
        # If there's no task available, navigate to the recorded page
        driver.get("https://au.agent.ai-media.community/agent-recorded")
        print("No task available, checking recorded page...")

# Close the chrome driver
driver.quit()
```
